```python

# Convert 'Amount_2021' and 'Amount_2022' columns to numeric
df['Amount_2021'] = df['Amount_2021'].str.replace(',', '').astype(float)
df['Amount_2022'] = df['Amount_2022'].str.replace(',', '').astype(float)

# Define a function for horizontal analysis
def horizontal_analysis(df):
    df['Change'] = df['Amount_2022'] - df['Amount_2021']
    df['Change %'] = ((df['Amount_2022'] - df['Amount_2021']) / df['Amount_2021']) * 100
    return df

# Define a function for vertical analysis
def vertical_analysis(df, base_item):
    df['% of ' + base_item] = (df['Amount_2022'] / df.loc[df['Category'] == base_item, 'Amount_2022'].values[0]) * 100
    return df
```

```python

# Perform horizontal analysis
df_horizontal = horizontal_analysis(df)

# Perform vertical analysis using 'Total operating expenses' as the base item
df_vertical = vertical_analysis(df_horizontal, 'Total operating expenses')
```

```python

# Printing horizontal analysis results
print("Horizontal Analysis:")
df_horizontal[['Category', 'Amount_2022', 'Amount_2021', 'Change', 'Change %']]
# print()
```

```python

# Printing vertical analysis results
print("Vertical Analysis (% of Total operating expenses):")
df_vertical[['Category', '% of Total operating expenses']]
```
