```python
import requests
from bs4 import BeautifulSoup

url = "https://www.annualreports.com/Companies?search=a"
comp_url = "https://www.annualreports.com/"
response = requests.get(url)

comp_urls = []
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    
    company_name_spans = soup.find_all('span', class_='companyName')
    
    hrefs = [span.a['href'] for span in company_name_spans if span.a]
    
    for href in hrefs:
        final_url = comp_url  + href
        comp_urls.append(final_url)
        print(final_url)
else:
    print("Failed to retrieve page. Status code:", response.status_code)
```

```python

import requests
from bs4 import BeautifulSoup
import pandas as pd
from concurrent.futures import ThreadPoolExecutor

data = pd.read_csv(r'company_urls.csv')
comp_urls = data['company_urls']
print(len(comp_urls))
print(comp_urls[0])

base_url = "https://www.annualreports.com/"
pdf_urls = []

def fetch_pdf_url(url_idx_tuple):
    url, idx = url_idx_tuple
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        view_btn_div = soup.find('div', class_='view_btn')
        final_pdf_url = 'EMPTY'
        if view_btn_div:
            a_tag = view_btn_div.find('a')
            if a_tag:
                href = a_tag.get('href')
                final_pdf_url = base_url + href
                pdf_urls.append((idx, url, final_pdf_url))
            else:
                pdf_urls.append((idx, url, 'NO_PDF'))
        else:
            pdf_urls.append((idx, url, 'NO_PDF'))
    else:
        pdf_urls.append((idx, url, '404_ERROR'))

    # Print a message indicating that the URL has been processed
    print(f"Processed URL {idx+1}/{len(comp_urls)}: {final_pdf_url}")

# Use ThreadPoolExecutor to fetch PDF URLs concurrently
with ThreadPoolExecutor() as executor:
    # Create a list of tuples containing URL and its index
    url_idx_tuples = [(url, idx) for idx, url in enumerate(comp_urls)]
    executor.map(fetch_pdf_url, url_idx_tuples)

# Sort pdf_urls based on the original sequence of comp_urls
pdf_urls.sort(key=lambda x: x[0])

print("PDF URLs fetched:", pdf_urls)
```
